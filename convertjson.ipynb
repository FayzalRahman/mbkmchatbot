```python
import pandas as pd
import json

# Load the Excel data into a DataFrame
excel_file = '../ppchatbot/jsonmbkmform.xlsx'
df = pd.read_excel(excel_file)

# Initialize an empty dictionary to store intents
intents = {}

# Iterate through the rows in the DataFrame
for index, row in df.iterrows():
    question = row['questions']
    answer = row['answers']

    if answer in intents:
        # If the response already exists, add the question to the existing patterns
        intents[answer]["patterns"].append(question)
    else:
        # Create a new intent for the response
        intents[answer] = {
            "tag": str(len(intents) + 1),
            "patterns": [question],
            "responses": [answer]
        }

# Create the JSON structure
json_data = {"intents": list(intents.values())}

# Convert to JSON format
json_str = json.dumps(json_data, indent=4)

# Print or save the JSON data
print(json_str)
```
